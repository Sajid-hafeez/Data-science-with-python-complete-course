## üîç What is Data Science?

- **Statistics & Mathematics** - for analysis and modeling

5. **Evaluation** - Assess model performance
6. **Deployment** - Implement the solution

## üõ†Ô∏è Hands-on Practice: Your First Data Science Code

Let's start with some basic Python operations that you'll use throughout the course:

### 1. Importing Essential Libraries

```python
# If you need to install these packages, uncomment and run these pip commands:
# pip install numpy pandas matplotlib seaborn
# pip install jupyter  # For Jupyter notebook support

# Import the fundamental libraries for data science
import numpy as np                 # Numerical computing
import pandas as pd                # Data manipulation
import matplotlib.pyplot as plt    # Basic plotting
import seaborn as sns             # Statistical visualization

# Set up plotting parameters
plt.style.use('seaborn-v0_8')     # Use seaborn style for better-looking plots
plt.rcParams['figure.figsize'] = (10, 6)  # Set default figure size

print("‚úÖ Libraries imported successfully!")
print(f"NumPy version: {np.__version__}")
print(f"Pandas version: {pd.__version__}")
```

### 2. Working with NumPy Arrays

```python
# Create a sample dataset using NumPy
np.random.seed(42)  # For reproducible results

# Generate sample data
sample_data = np.random.normal(50, 15, 1000)  # 1000 values, mean=50, std=15

print("Sample Statistics:")
print(f"Mean: {np.mean(sample_data):.2f}")
print(f"Standard Deviation: {np.std(sample_data):.2f}")
print(f"Min: {np.min(sample_data):.2f}")
print(f"Max: {np.max(sample_data):.2f}")
print(f"Shape: {sample_data.shape}")
```

### 3. Creating Your First DataFrame

```python
# Load a real dataset from GitHub - Titanic dataset
# This is one of the most famous datasets in data science for learning

# URL to the Titanic dataset on GitHub
url = 'https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv'

try:
    # Load the dataset directly from GitHub
    df = pd.read_csv(url)
    
    print("‚úÖ Dataset loaded successfully from GitHub!")
    print(f"Dataset: Titanic Passenger Data")
    print(f"Shape: {df.shape}")
    print(f"Columns: {list(df.columns)}")
    
    print("\n" + "="*50)
    print("First 5 rows:")
    display(df.head())
    
    print("\n" + "="*50)
    print("Dataset Description:")
    print("This dataset contains information about Titanic passengers including:")
    print("- PassengerId: Unique ID for each passenger")
    print("- Survived: Whether passenger survived (0=No, 1=Yes)")
    print("- Pclass: Passenger class (1st, 2nd, 3rd)")
    print("- Name: Passenger name")
    print("- Sex: Gender")
    print("- Age: Age in years")
    print("- SibSp: Number of siblings/spouses aboard")
    print("- Parch: Number of parents/children aboard")
    print("- Ticket: Ticket number")
    print("- Fare: Passenger fare")
    print("- Cabin: Cabin number")
    print("- Embarked: Port of embarkation (C=Cherbourg, Q=Queenstown, S=Southampton)")
    
except Exception as e:
    print(f"‚ùå Error loading dataset: {e}")
    print("Please check your internet connection or try again later.")
    
    # Fallback: Create a simple synthetic dataset if GitHub is not accessible
    print("\nüîÑ Creating fallback synthetic dataset...")
    np.random.seed(42)
    n_passengers = 100
    
    fallback_data = {
        'PassengerId': range(1, n_passengers + 1),
        'Survived': np.random.choice([0, 1], n_passengers, p=[0.6, 0.4]),
        'Pclass': np.random.choice([1, 2, 3], n_passengers, p=[0.2, 0.3, 0.5]),
        'Sex': np.random.choice(['male', 'female'], n_passengers),
        'Age': np.random.normal(30, 12, n_passengers).round(1),
        'Fare': np.random.exponential(30, n_passengers).round(2)
    }
    
    df = pd.DataFrame(fallback_data)
    df['Age'] = df['Age'].clip(lower=0, upper=80)  # Realistic age range
    
    print("‚úÖ Fallback dataset created successfully!")
    print(f"Shape: {df.shape}")
    display(df.head())
```

### 4. Basic Data Exploration

```python
# Get basic information about the Titanic dataset
print("Dataset Info:")
print(df.info())

print("\n" + "="*50)
print("Summary Statistics for Numerical Columns:")
display(df.describe())

print("\n" + "="*30)
print("Survival Rate:")
if 'Survived' in df.columns:
    survival_rate = df['Survived'].mean()
    print(f"Overall survival rate: {survival_rate:.1%}")
    print(f"Passengers who survived: {df['Survived'].sum()}")
    print(f"Passengers who didn't survive: {len(df) - df['Survived'].sum()}")

print("\n" + "="*30)
print("Passenger Class Distribution:")
if 'Pclass' in df.columns:
    pclass_counts = df['Pclass'].value_counts().sort_index()
    print(pclass_counts)

print("\n" + "="*30)
print("Gender Distribution:")
if 'Sex' in df.columns:
    gender_counts = df['Sex'].value_counts()
    print(gender_counts)
```

```python
# Check for missing values
print("Missing Values Count:")
missing_values = df.isnull().sum()
print(missing_values)

print("\n" + "="*40)
print("Missing Values Percentage:")
missing_percentage = (df.isnull().sum() / len(df) * 100).round(2)
missing_df = pd.DataFrame({
    'Missing Count': missing_values,
    'Missing Percentage': missing_percentage
})
print(missing_df[missing_df['Missing Count'] > 0])

print("\n" + "="*30)
print("Data Types:")
print(df.dtypes)

print("\n" + "="*30)
print("Key Observations about Missing Data:")
if 'Age' in df.columns and df['Age'].isnull().sum() > 0:
    print(f"- Age has {df['Age'].isnull().sum()} missing values ({df['Age'].isnull().sum()/len(df)*100:.1f}%)")
if 'Cabin' in df.columns and df['Cabin'].isnull().sum() > 0:
    print(f"- Cabin has {df['Cabin'].isnull().sum()} missing values ({df['Cabin'].isnull().sum()/len(df)*100:.1f}%)")
if 'Embarked' in df.columns and df['Embarked'].isnull().sum() > 0:
    print(f"- Embarked has {df['Embarked'].isnull().sum()} missing values ({df['Embarked'].isnull().sum()/len(df)*100:.1f}%)")
print("- Missing data is common in real-world datasets and requires careful handling!")
```

### 5. Your First Data Visualization

```python
# Create visualizations for the Titanic dataset
plt.figure(figsize=(15, 10))

# Subplot 1: Age Distribution
plt.subplot(2, 3, 1)
if 'Age' in df.columns:
    # Remove missing values for visualization
    age_data = df['Age'].dropna()
    plt.hist(age_data, bins=20, color='skyblue', alpha=0.7, edgecolor='black')
    plt.title('Distribution of Passenger Ages')
    plt.xlabel('Age')
    plt.ylabel('Frequency')

# Subplot 2: Fare Distribution
plt.subplot(2, 3, 2)
if 'Fare' in df.columns:
    fare_data = df['Fare'].dropna()
    plt.hist(fare_data, bins=20, color='lightgreen', alpha=0.7, edgecolor='black')
    plt.title('Distribution of Ticket Fares')
    plt.xlabel('Fare')
    plt.ylabel('Frequency')

# Subplot 3: Passenger Class Distribution
plt.subplot(2, 3, 3)
if 'Pclass' in df.columns:
    pclass_counts = df['Pclass'].value_counts().sort_index()
    plt.bar(pclass_counts.index, pclass_counts.values, color='salmon', alpha=0.7)
    plt.title('Passengers by Class')
    plt.xlabel('Passenger Class')
    plt.ylabel('Number of Passengers')
    plt.xticks([1, 2, 3], ['1st Class', '2nd Class', '3rd Class'])

# Subplot 4: Gender Distribution
plt.subplot(2, 3, 4)
if 'Sex' in df.columns:
    gender_counts = df['Sex'].value_counts()
    plt.pie(gender_counts.values, labels=gender_counts.index, autopct='%1.1f%%', 
            colors=['lightblue', 'pink'], startangle=90)
    plt.title('Gender Distribution')

# Subplot 5: Survival Rate
plt.subplot(2, 3, 5)
if 'Survived' in df.columns:
    survival_counts = df['Survived'].value_counts()
    labels = ['Did not survive', 'Survived']
    plt.pie(survival_counts.values, labels=labels, autopct='%1.1f%%', 
            colors=['lightcoral', 'lightgreen'], startangle=90)
    plt.title('Survival Rate')

# Subplot 6: Embarkation Port
plt.subplot(2, 3, 6)
if 'Embarked' in df.columns:
    embarked_counts = df['Embarked'].value_counts()
    plt.bar(embarked_counts.index, embarked_counts.values, color='gold', alpha=0.7)
    plt.title('Embarkation Ports')
    plt.xlabel('Port')
    plt.ylabel('Number of Passengers')
    # Add port names for clarity
    port_names = {'S': 'Southampton', 'C': 'Cherbourg', 'Q': 'Queenstown'}
    current_labels = plt.gca().get_xticklabels()
    new_labels = [port_names.get(label.get_text(), label.get_text()) for label in current_labels]
    plt.gca().set_xticklabels(new_labels, rotation=45)

plt.tight_layout()
plt.show()

print("üéâ Congratulations! You've created your first real-world data visualizations!")
print("üìä These charts show the distribution of various features in the Titanic dataset.")
print("üí° Notice how real data often has interesting patterns and missing values!")
```

### 6. Basic Data Analysis

```python
# Analyze survival patterns in the Titanic dataset
print("Survival Analysis:")

if 'Survived' in df.columns and 'Sex' in df.columns:
    print("\n1. Survival Rate by Gender:")
    gender_survival = df.groupby('Sex')['Survived'].agg(['count', 'sum', 'mean']).round(3)
    gender_survival.columns = ['Total', 'Survived', 'Survival_Rate']
    print(gender_survival)

if 'Survived' in df.columns and 'Pclass' in df.columns:
    print("\n2. Survival Rate by Passenger Class:")
    class_survival = df.groupby('Pclass')['Survived'].agg(['count', 'sum', 'mean']).round(3)
    class_survival.columns = ['Total', 'Survived', 'Survival_Rate']
    class_survival.index = ['1st Class', '2nd Class', '3rd Class']
    print(class_survival)

if 'Age' in df.columns:
    print("\n3. Age Statistics by Survival:")
    age_survival = df.groupby('Survived')['Age'].agg(['count', 'mean', 'median', 'std']).round(2)
    age_survival.index = ['Did not survive', 'Survived']
    print(age_survival)

if 'Fare' in df.columns and 'Survived' in df.columns:
    print("\n4. Fare Statistics by Survival:")
    fare_survival = df.groupby('Survived')['Fare'].agg(['count', 'mean', 'median', 'std']).round(2)
    fare_survival.index = ['Did not survive', 'Survived']
    print(fare_survival)

print("\n" + "="*50)
print("Key Insights:")
if 'Sex' in df.columns and 'Survived' in df.columns:
    female_survival = df[df['Sex'] == 'female']['Survived'].mean()
    male_survival = df[df['Sex'] == 'male']['Survived'].mean()
    print(f"‚Ä¢ Women had a {female_survival:.1%} survival rate vs {male_survival:.1%} for men")

if 'Pclass' in df.columns and 'Survived' in df.columns:
    first_class_survival = df[df['Pclass'] == 1]['Survived'].mean()
    third_class_survival = df[df['Pclass'] == 3]['Survived'].mean()
    print(f"‚Ä¢ 1st class passengers: {first_class_survival:.1%} survival vs 3rd class: {third_class_survival:.1%}")

print("‚Ä¢ This demonstrates how social factors affected survival during the disaster")
```

### 7. Correlation Analysis

```python
# Calculate correlation matrix for numerical columns in Titanic dataset
numerical_columns = df.select_dtypes(include=[np.number]).columns.tolist()

# Remove PassengerId as it's just an identifier
if 'PassengerId' in numerical_columns:
    numerical_columns.remove('PassengerId')

if len(numerical_columns) >= 2:
    correlation_matrix = df[numerical_columns].corr()
    
    # Create a heatmap
    plt.figure(figsize=(10, 8))
    sns.heatmap(correlation_matrix, 
                annot=True, 
                cmap='coolwarm', 
                center=0,
                square=True,
                fmt='.3f',
                cbar_kws={'label': 'Correlation Coefficient'})
    plt.title('Correlation Matrix: Titanic Dataset Features', fontsize=14, fontweight='bold')
    plt.tight_layout()
    plt.show()
    
    print("Insights from Correlation Analysis:")
    
    # Survival correlations
    if 'Survived' in correlation_matrix.columns:
        print("\nCorrelations with Survival:")
        survival_corr = correlation_matrix['Survived'].drop('Survived').sort_values(key=abs, ascending=False)
        for feature, corr in survival_corr.items():
            direction = "positively" if corr > 0 else "negatively"
            strength = "strong" if abs(corr) > 0.3 else "moderate" if abs(corr) > 0.1 else "weak"
            print(f"‚Ä¢ {feature}: {corr:.3f} ({strength} {direction} correlated)")
    
    # Other interesting correlations
    print(f"\nOther Notable Correlations:")
    if 'Fare' in correlation_matrix.columns and 'Pclass' in correlation_matrix.columns:
        fare_class_corr = correlation_matrix.loc['Fare', 'Pclass']
        print(f"‚Ä¢ Fare and Passenger Class: {fare_class_corr:.3f}")
        print("  (Higher class passengers paid more, as expected)")
    
    if 'Age' in correlation_matrix.columns and 'Fare' in correlation_matrix.columns:
        age_fare_corr = correlation_matrix.loc['Age', 'Fare']
        print(f"‚Ä¢ Age and Fare: {age_fare_corr:.3f}")
    
else:
    print("Not enough numerical columns for correlation analysis.")
    print(f"Available numerical columns: {numerical_columns}")

print(f"\nüìä Correlation helps us understand relationships between variables!")
print("üîç Values close to +1 or -1 indicate strong relationships.")
```

### 8. Creating a Simple Scatter Plot

```python
# Create scatter plots to explore relationships in the Titanic dataset
plt.figure(figsize=(15, 5))

# Scatter plot 1: Age vs Fare (colored by survival)
plt.subplot(1, 3, 1)
if 'Age' in df.columns and 'Fare' in df.columns and 'Survived' in df.columns:
    # Remove rows with missing Age or Fare values
    clean_data = df.dropna(subset=['Age', 'Fare'])
    
    # Create scatter plot with different colors for survivors vs non-survivors
    survivors = clean_data[clean_data['Survived'] == 1]
    non_survivors = clean_data[clean_data['Survived'] == 0]
    
    plt.scatter(non_survivors['Age'], non_survivors['Fare'], 
               alpha=0.6, c='red', label='Did not survive', s=30)
    plt.scatter(survivors['Age'], survivors['Fare'], 
               alpha=0.6, c='green', label='Survived', s=30)
    
    plt.xlabel('Age')
    plt.ylabel('Fare')
    plt.title('Age vs Fare by Survival')
    plt.legend()
    plt.grid(True, alpha=0.3)

# Scatter plot 2: Age vs Fare (colored by passenger class)
plt.subplot(1, 3, 2)
if 'Age' in df.columns and 'Fare' in df.columns and 'Pclass' in df.columns:
    clean_data = df.dropna(subset=['Age', 'Fare'])
    
    colors = ['red', 'blue', 'green']
    class_labels = ['1st Class', '2nd Class', '3rd Class']
    
    for i, pclass in enumerate([1, 2, 3]):
        class_data = clean_data[clean_data['Pclass'] == pclass]
        plt.scatter(class_data['Age'], class_data['Fare'], 
                   alpha=0.6, c=colors[i], label=class_labels[i], s=30)
    
    plt.xlabel('Age')
    plt.ylabel('Fare')
    plt.title('Age vs Fare by Passenger Class')
    plt.legend()
    plt.grid(True, alpha=0.3)

# Scatter plot 3: Family size vs Survival rate
plt.subplot(1, 3, 3)
if 'SibSp' in df.columns and 'Parch' in df.columns and 'Survived' in df.columns:
    # Create family size feature
    df_temp = df.copy()
    df_temp['FamilySize'] = df_temp['SibSp'] + df_temp['Parch'] + 1  # +1 for the passenger themselves
    
    # Calculate survival rate by family size
    family_survival = df_temp.groupby('FamilySize')['Survived'].agg(['count', 'mean']).reset_index()
    family_survival = family_survival[family_survival['count'] >= 5]  # Only show family sizes with 5+ passengers
    
    plt.scatter(family_survival['FamilySize'], family_survival['mean'], 
               s=family_survival['count']*10, alpha=0.7, c='purple')
    
    plt.xlabel('Family Size')
    plt.ylabel('Survival Rate')
    plt.title('Family Size vs Survival Rate')
    plt.grid(True, alpha=0.3)
    
    # Add text annotations for data points
    for _, row in family_survival.iterrows():
        plt.annotate(f'n={row["count"]}', 
                    (row['FamilySize'], row['mean']), 
                    xytext=(5, 5), textcoords='offset points', fontsize=8)

plt.tight_layout()
plt.show()

print("üìä These scatter plots reveal interesting patterns in the Titanic dataset!")
print("üîç Key observations:")
print("‚Ä¢ First plot: Notice how survival varies with age and fare")
print("‚Ä¢ Second plot: Higher class passengers generally paid more")
print("‚Ä¢ Third plot: Family size affected survival chances")
print("üí° Real-world data often reveals unexpected relationships!")
```
